```python
devices = tuyapower.deviceScan()
for ip in devices:
    id = devices[ip]['gwId']
    key = devices[ip]['productKey']
    vers = devices[ip]['version']
    (on, w, mA, V, err) = tuyapower.deviceInfo(id, ip, key, vers)
    print("Device at %s: ID %s, state=%s, W=%s, mA=%s, V=%s [%s]"%(ip,id,on,w,mA,V,err))
```
